# 資料分析工具入門：pandas + matplotlib

這份筆記本教兩個之後一定會用到的工具：
- **pandas**：處理表格資料（像程式版的 Excel）
- **matplotlib**：把資料畫成圖表

1. pandas 基礎：DataFrame 是什麼、怎麼看資料

## pandas 基礎：DataFrame 是什麼

pandas 最核心的東西叫 **DataFrame**，你可以想成「一張表格」：
有欄位（直的，像 Excel 的欄）、有一列一列的資料（橫的）。
